# Introduction to Stochastic Gradient Descent (SGD)

Stochastic Gradient Descent (SGD) is a powerful optimization algorithm often used in training a wide range of machine learning models, particularly in the field of deep learning. SGD is an iterative method for optimizing an objective function, typically a loss function that is defined as the sum of losses over individual examples.

## Fundamentals of SGD

SGD updates the parameters of the model incrementally, one training example (or a small batch) at a time, which is what distinguishes it from batch gradient descent, where updates are made using the whole dataset. This stochastic nature often leads to faster convergence because the model parameters are updated more frequently. At the end, we strive for finding the global minimum on the loss manifold. The challenge in deep learning is to not remain in some bad local minimum, which does not represent well the data.

![](https://d2l.ai/_images/output_gd_79c039_87_1.svg)

### Mathematical Representation

The parameters are updated according to the rule:


$\mathbf{w} = \mathbf{w} - \eta \nabla_\mathbf{w} L(\mathbf{\hat{y}}^{(i)}, \mathbf{y}^{(i)})$

Where:
* $\mathbf{w}$ represents the parameters of the model (weights + bias).
* $\eta$ is the learning rate, controlling the step size during the minimization process. Setting it to a very small value will lead to very precise results, but convergence is very slow. If we set it to an high value, convergence is faster, but also very rough or bad. Hence, we also search for an optimum here. Later, we will also discuss possibilities to adapt the learning rate dynamically.
* $L(\mathbf{\hat{y}}^{(i)}, \mathbf{y}^{(i)})$ is the cost function evaluated on a single data sample
* $\nabla_\mathbf{w}$ is the gradient of the cost function with respect to the parameters.

**Benefits of SGD**
* Efficiency: SGD is computationally less intensive, which makes it particularly useful for large datasets.
* Frequent updates: Frequent updates with a high variance cause the objective function to fluctuate heavily, which can help escape local minima.

**Difference of Gradient Descent and Stochastic Gradient Descent**

Training on all data at once is called gradient descent. But, as this might be quite costly for a large amount of data we approximate GD by using only one example at a time in stochastic gradient descent. As this is the other extremum and getting quite slow, the optimum is between both and is called `Minibatch-Stochastic-Gradient-Descent`. We take a mini-batch of size *B* and calculate the gradient on this batch. This is quite helpful as the batches add some kind of noise, which will help finding better optima.

**Variants of SGD**

Several variants of SGD introduce mechanisms to adapt the learning rate during training or to take into account the previous gradients to smooth the learning process. Some popular variants include:

* **Momentum**: Helps accelerate SGD in the relevant direction and dampens oscillations.
* **Adagrad**: Adapts the learning rate to the parameters, performing larger updates for infrequent parameters.
* **Adam**: Combines the advantages of Adagrad and RMSprop, another adaptive learning rate method.

Each variant provides certain benefits in specific scenarios, depending on the nature of the problem and the data.

## BONUS - Adam: Adaptive Moment Estimation
Adam is the most widely used optimization algorithm in deep learning. It combines the advantages of two other extensions of stochastic gradient descent. Specifically:

* Adaptive Gradient Algorithm (AdaGrad) that maintains a per-parameter learning rate that improves performance on problems with sparse gradients.
* Root Mean Square Propagation (RMSProp) that also maintains per-parameter learning rates that are adapted based on the average of recent magnitudes of the gradients for the weight.

Both approaches are not mentioned here as Adam usually outperforms both of them. If you like to read further about both approaches, please refer to [RMSProp](https://d2l.ai/chapter_optimization/rmsprop.html#rmsprop) and [AdaGrad](https://d2l.ai/chapter_optimization/adagrad.html).

### How Momentum Works
Momentum helps the gradient descent algorithm to navigate along the relevant directions and smooths out the updates. It does this by adding a fraction $\gamma$ of the update vector of the past time step to the current update vector:

$v_t = \gamma v_{t-1} + \eta \nabla_\mathbf{w}L(\mathbf{w})$

$\mathbf{w} = \mathbf{w} - v_t$

$\gamma$ is typically set to 0.9 or a similar value.
$\eta$ is still the learning rate.
$\nabla_\mathbf{w}L(\mathbf{w})$ is the gradient of the loss function with respect to the parameters $\mathbf{w}$

In Python you could add a momentum by passing the corresponding parameter to the SGD function.

`optimizer = torch.optim.SGD([params], lr=0.1, momentum=0.9)`

### How Adam Works
Adam computes adaptive learning rates for each parameter. In addition to storing an exponentially decaying average of past squared gradients $v_t$ like RMSprop, Adam also keeps an exponentially decaying average of past gradients $m_t$ , similar to momentum:

\begin{align}
m_t &= \beta_1 m_{t-1} + (1 - \beta_1) \nabla_{\mathbf{w}}L(\mathbf{w}) \\
v_t &= \beta_2 v_{t-1} + (1 - \beta_2) (\nabla_{\mathbf{w}}L(\mathbf{w}))^2 \\
\hat{m}_t &= \frac{m_t}{1 - \beta_1^t} \\
\hat{v}_t &= \frac{v_t}{1 - \beta_2^t} \\
\mathbf{w} &= \mathbf{w} - \frac{\eta}{\sqrt{\hat{v}_t} + \epsilon} \hat{m}_t \\
\end{align}

$β_1$ and $β_2$ are estimates of the decay rates — typically taken as 0.9 and 0.999 respectively.

$\epsilon$ is a very small number to prevent any division by zero in the implementation.

```python
import torch

# Example parameters and gradient
params = torch.tensor([1.0, 2.0], requires_grad=True)
# Instead of calling SGD, we call Adam directly.
optimizer = torch.optim.Adam([params], lr=0.001)

# Example loss function and backward pass
loss = params.norm()
loss.backward()

# Step through optimizer to update params
optimizer.step()
```

## How Gradient Descent Works in detail and with more layers

So far, we only optimized simple functions. In order to use SGD in deep learning, we need to understand how it works also for subsequent layers in a neural network.

First, again the general workflow, which is also correct for Deep Learning.

1. **Objective Function**:
   The objective or cost function \( $J(\theta)$ \) quantifies the error between the predicted outputs of the model and the actual data. \( $\theta$ \) represents the model parameters (weights and biases).

2. **Gradient Calculation**:
   Gradient descent updates the model parameters by moving them in the direction that reduces the cost function the fastest. This direction is the negative gradient of the cost function with respect to the parameters:
   
   $\nabla_{\theta} J(\theta) = \left( \frac{\partial J}{\partial \theta_1}, \frac{\partial J}{\partial \theta_2}, \ldots, \frac{\partial J}{\partial \theta_n} \right)$

   Here, each component of the gradient vector represents the partial derivative of the cost function with respect to each parameter. This is the crucial part for deep learning. Out network consists of several layers, which each several nodes and parameters. The idea is to calculate the partial derivative of the Loss function relative to each parameter. This sounds weird, but each parameter has a contribution to the total Loss. Hence, it is possible to derive the partial derivative for each of them. Using this partial derivatives, we can calculate each update for each parameter.

3. **Parameter Update**:
   Parameters are updated by taking a step proportional to the negative gradient:
   
   $\theta = \theta - \eta \nabla_{\theta} J(\theta)$
   
   where \( $\eta$ \) is the learning rate a scalar that determines the size of the step to take on each iteration. A higher learning rate makes the steps bigger, allowing the model to learn faster but potentially overshooting minima; a smaller learning rate makes the learning process steadier but slower.

## Backpropagation and Error Propagation

1. **Output Error**:
   The process begins at the output layer, where the error derivative is first calculated. This is often straightforward, especially with common loss functions like mean squared error for regression or cross-entropy for classification:

   * For mean squared error: $\frac{\partial J}{\partial \hat{y}} = \hat{y} - y$
   * For cross-entropy in a binary classification with a sigmoid output: $\frac{\partial J}{\partial z^{(L)}} = \hat{y} - y$

   Here, $\hat{y}$ is the predicted output, $y$ is the actual output, and $z^{(L)}$ is the input to the output layer.

2. **Hidden Layers**:
   The derivative of the error is propagated backward through each hidden layer. For each layer \( L \), the error derivative with respect to the output of layer \( L \) is used to compute the error derivative with respect to the input of layer $L+1$, which is then used to compute the error derivatives with respect to the weights and biases:
   
   $\frac{\partial J}{\partial z^{(L)}} = \left( \frac{\partial J}{\partial a^{(L+1)}} \right) \cdot \left( \frac{\partial a^{(L+1)}}{\partial z^{(L)}} \right)$
   
   $\frac{\partial J}{\partial W^{(L)}} = \left( \frac{\partial J}{\partial z^{(L)}} \right) \cdot \left( \frac{\partial z^{(L)}}{\partial W^{(L)}} \right)$

3. **Gradient Descent Update**:
   After computing these gradients, the weights and biases of each layer are updated using gradient descent.
